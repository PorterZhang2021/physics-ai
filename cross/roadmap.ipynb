## Part 2: Six Layers in Detail

### Layer 1: Mathematical Foundations

**Why it is the foundation:** Quantum mechanics is written in linear algebra; topological invariants are differential-geometry integrals; symmetry classification uses group theory. Without solid math, physics becomes rote memorization, not true understanding.

| Module | Specific Content | Textbook | Priority |
|------|----------|------|----------|
| Linear Algebra | Matrix operations, eigenvalues, SVD, unitary and Hermitian matrices | Strang《Linear Algebra》| ★★★★★ |
| Calculus & Complex Analysis | Chain rule, multivariate partial derivatives, complex numbers, analytic functions | Review engineering textbooks | ★★★★★ |
| Probability | Conditional probability, random variables, normal distribution, Bayes | Blitzstein《Probability》| ★★★★☆ |
| Differential Geometry (intro) | Manifold intuition, fiber bundles (physical intuition only)| Nakahara (selective) | ★★★★☆ |
| Group Theory (basics) | Group definitions, representation theory, crystal point groups | Dresselhaus (selective) | ★★★☆☆ |

**Study tips:** Start linear algebra and calculus immediately. Postpone differential geometry and group theory until before Layer 3; do not tackle them from the start.

---

### Layer 2: Core Physics

**Why it matters:** Quantum mechanics is the universal language; solid-state physics is the stage; statistical mechanics explains finite-temperature effects. Missing any one leaves research incomplete.

| Module | Textbook | Required Chapters | Priority |
|------|------|----------|----------|
| Quantum Mechanics | Griffiths *Introduction to Quantum Mechanics*| **Ch.1-5 fully, do exercises** | ★★★★★ |
| Solid State Physics | Kittel *Introduction to Solid State Physics*| Ch.1-2 (crystals), Ch.7-8 (bands), Ch.10 (Fermi gas)| ★★★★★ |
| Statistical Mechanics | Huang *Statistical Mechanics* or Pathria | Basic thermal statistics, Fermi-Dirac distribution | ★★★★☆ |

**Key milestones:** Can independently derive stationary states for infinite well and harmonic oscillator; can explain why solids have band structures。

---

### Layer 3: Specialized Physics (The Core Knowledge of This Direction)

**This layer is your main research arena.**

| Module | Textbook | Core Concepts | Priority |
|------|------|----------|----------|
| Berry Phase & Topological Invariants | Vanderbilt *Berry Phases* Ch.1-3 + Berry 1984 paper | Berry connection, Berry curvature, Chern number | ★★★★★ |
| Topological Insulators & Superconductors | Asbóth *Short Course* full (intro) → Bernevig *Topological Insulators* (advanced)| SSH model, bulk-boundary correspondence, Z₂ classification, ten-fold symmetry | ★★★★★ |
| Quantum Transport Theory | Datta《Electronic Transport》Ch.1-4 | Landauer formula, Büttiker multi-terminal formula, scattering matrix | ★★★★★ |
| Quantum Many-Body (advanced)| Altland & Simons *Condensed Matter Field Theory* Ch.1-2 | Second quantization, field-theory language | ★★★★☆ |

**Required review papers (read alongside textbooks to build field vision):**
- Hasan & Kane, Rev. Mod. Phys. 2010 — Foundational review of topological insulators
- Qi & Zhang, Rev. Mod. Phys. 2011 — More comprehensive review
- Beenakker, Rev. Mod. Phys. 1997 — Quantum transport review
