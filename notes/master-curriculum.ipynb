### 如何读论文（关键技能）

```
第一遍（30 分钟）：只看 Abstract + Introduction + 结论图
    目标：理解这篇论文解决了什么问题、结果是什么

第二遍（2-3 小时）：看方法 + 主要实验
    目标：理解他们怎么做的

第三遍（若干天）：推导公式 + 尝试复现
    目标：真正理解，发现细节
```

> 这个方向是在完成「阶段四：量子力学」之后，沿 **C2 路线**深入的一个专项。
> 它是当前凝聚态物理 + AI 结合最紧密的领域之一。

## 先理解这个方向在研究什么

**拓扑物态（Topological Matter）：**
```
经典的物质分类：导体 / 绝缘体 / 半导体
  → 用能带结构区分（有没有带隙）

拓扑物态的发现（2000年代起）：
  → 有些绝缘体，内部是绝缘的，但表面/边缘必然存在导电通道
  → 这个导电通道受「拓扑保护」，不会被杂质或缺陷破坏
  → 原因不在能带高低，而在能带的「拓扑形状」
```

**量子输运（Quantum Transport）：**
```
经典输运：欧姆定律 R = ρL/A，电阻随长度线性增大

量子输运（纳米尺度/低温）：
  → 电子保持量子相干性，干涉效应出现
  → 量子霍尔效应：横向电导精确量子化（e²/h 的整数倍）
  → 弹道输运：电子不散射地穿过，电导不随长度变化
  → 拓扑边缘态的输运是无耗散的
```

**为什么和 AI 结合：**
```
1. 相变探测：拓扑相变很难用传统方法识别 → 用 ML 分类
2. 拓扑不变量计算：复杂系统的陈数、Z₂不变量 → 用神经网络计算
3. 输运性质预测：Green 函数计算昂贵 → 用 ML 加速
4. 材料设计：预测哪些材料是拓扑绝缘体 → 用 ML 筛选
```

**2016 年诺贝尔物理学奖** 就颁给了拓扑相变和物质拓扑相领域。

````python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 直观演示拓扑的核心思想：连续变形不能改变的性质
fig, axes = plt.subplots(1, 3, figsize=(13, 4))

theta = np.linspace(0, 2*np.pi, 300)

# 圆（亏格=0，无洞）
axes[0].plot(np.cos(theta), np.sin(theta), 'steelblue', lw=3)
axes[0].plot(np.cos(theta)*0.5+1.5, np.sin(theta)*0.3, 'coral', lw=3, linestyle='--')
axes[0].set_xlim(-2, 3)
axes[0].set_ylim(-1.5, 1.5)
axes[0].set_aspect('equal')
axes[0].set_title('拓扑等价\n（圆可以连续变成椭圆）', fontsize=10)
axes[0].annotate('', xy=(1.5, 0), xytext=(0.7, 0),
                arrowprops=dict(arrowstyle='->', color='green', lw=2))
axes[0].text(1.0, 0.15, '连续变形', fontsize=9, color='green')
axes[0].axis('off')

# 环面（亏格=1，有洞）
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, 2*np.pi, 100)
U, V = np.meshgrid(u, v)
R, r = 1.0, 0.4
X = (R + r*np.cos(V)) * np.cos(U)
Y = (R + r*np.cos(V)) * np.sin(U)
Z = r * np.sin(V)
ax3d = fig.add_subplot(132, projection='3d') if False else axes[1]
axes[1].contourf(X, Y, Z, levels=15, cmap='Blues', alpha=0.7)
axes[1].set_title('拓扑不等价\n（环面有洞，无法变成球）', fontsize=10)
axes[1].text(0, 0, '洞=拓扑不变量\n（陈数、绕数）', ha='center', va='center',
            fontsize=9, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
axes[1].axis('off')

# 拓扑绝缘体的边缘态示意
x = np.linspace(-3, 3, 200)
E_bulk_upper =  1.0 + 0.3*x**2
E_bulk_lower = -1.0 - 0.3*x**2
E_edge_right =  0.6 * x
E_edge_left  = -0.6 * x

axes[2].fill_between(x, E_bulk_lower, -1.8, alpha=0.15, color='steelblue', label='体态（bulk）')
axes[2].fill_between(x, E_bulk_upper,  1.8, alpha=0.15, color='steelblue')
axes[2].plot(x, E_bulk_upper, 'steelblue', lw=2)
axes[2].plot(x, E_bulk_lower, 'steelblue', lw=2)
axes[2].plot(x, E_edge_right, 'r-', lw=2.5, label='拓扑边缘态')
axes[2].plot(x, E_edge_left, 'r--', lw=2.5)
axes[2].axhline(0, color='gray', lw=0.8, linestyle=':')
axes[2].fill_betweenx([-0.9, 0.9], -3, 3, alpha=0.05, color='green')
axes[2].text(0, 0, '带隙', ha='center', va='center', fontsize=9, color='green')
axes[2].set_xlabel('动量 k')
axes[2].set_ylabel('能量 E')
axes[2].set_title('拓扑绝缘体能带结构\n（内部有带隙，边缘有导电态）', fontsize=10)
axes[2].legend(fontsize=8)
axes[2].set_ylim(-1.8, 1.8)

plt.suptitle('拓扑的核心思想：不能连续变形消除的性质', fontsize=12)
plt.tight_layout()
plt.show()
````

### 补充二：Berry 相位（约 3 周）

**为什么重要：** Berry 相位是拓扑物理的数学核心。拓扑不变量（陈数）本质上是 Berry 曲率的积分。

**学习资源（按顺序）：**

| 资源 | 内容 | 说明 |
|------|------|------|
| Griffiths Ch.10（绝热近似）| 为 Berry 相位做准备 | 先把这章搞懂 |
| Berry (1984) 原始论文 | Quantal Phases in Adiabatic Processes | 3 页，非常清晰 |
| David Vanderbilt《Berry Phases in Electronic Structure Theory》Ch.1-2 | Berry 相位的现代处理 | 免费 PDF |

**核心公式（要能理解每个符号的含义）：**
```
Berry 联络：  A(k) = i<u(k)|∇_k|u(k)>
Berry 曲率：  Ω(k) = ∇_k × A(k)
陈数（Chern Number）：C = (1/2π) ∫ Ω(k) d²k

陈数是整数，是拓扑不变量——连续变形不能改变它
量子霍尔效应的霍尔电导 = C × e²/h
```

**核心公式（Landauer 公式——量子输运的基石）：**
```
电导 G = (e²/h) × T

T 是电子从左端到右端的透射概率
当 T=1（完美透射），G = e²/h ≈ 77.5 μS

这个量子化的电导单位在实验上可以精确测量
拓扑边缘态的输运就是 T=1 的理想情况
```

## AI × 拓扑物态的具体结合点

这是你的最终目标区域。以下是几个具体的研究方向和对应的核心论文。

---

### 结合点一：ML 识别拓扑相变

**问题：** 拓扑相变无法用局域序参量描述（朗道理论不适用），用眼睛看数据很难判断。  
**AI 方案：** 把不同参数下的量子态/能带输入神经网络，让它学会分类。

**入门论文：**
- Carrasquilla & Melko, *Nat. Phys.* 13, 431 (2017) — ML 识别物理相变的开创性工作
- Rodriguez-Nieva & Scheurer, *Nat. Phys.* 15, 790 (2019) — 无监督学习识别拓扑相
- Zhang et al., *Phys. Rev. B* 99, 054209 (2019) — 专门针对拓扑相分类

---

### 结合点二：神经网络计算拓扑不变量

**问题：** 对复杂材料计算陈数、Z₂ 不变量计算量很大，需要遍历布里渊区。  
**AI 方案：** 训练神经网络，输入哈密顿量参数，直接预测拓扑不变量。

**入门论文：**
- Scheurer & Slager, *Phys. Rev. Lett.* 124, 226401 (2020)
- Holanda & Vargas-Paredes, *Phys. Rev. B* (2020)

---

### 结合点三：ML 加速量子输运计算

**问题：** NEGF 方法计算输运性质（电导、透射谱）计算量是 O(N³)。  
**AI 方案：** 用 GNN（图神经网络）或 Transformer 预测输运性质。

**入门论文：**
- Ghosh et al., *Phys. Rev. B* 107, 235417 (2023)
- 相关方向：AI for DFT（密度泛函理论）也有大量工作

---

### 结合点四：对称性约束的神经网络

**问题：** 拓扑物态由对称性保护，普通神经网络不自动尊重这些对称性。  
**AI 方案：** 设计等变神经网络（Equivariant NN），把时间反演、空间反演等对称性硬编码进网络结构。

**入门论文：**
- 这与 DeepMind 的 AlphaFold 2 使用的等变结构同源
- Batzner et al., *Nat. Commun.* 13, 2453 (2022) — NequIP 方法
- 搜索关键词：equivariant neural network + topological

````python
# 演示 SSH 模型（最简单的拓扑模型，Asbóth 书第一章）
# SSH = Su-Schrieffer-Heeger 模型，一维拓扑绝缘体原型

def ssh_hamiltonian(v, w, N):
    """构建 SSH 模型的哈密顿量
    v: 胞内跳跃强度, w: 胞间跳跃强度, N: 单胞数"""
    H = np.zeros((2*N, 2*N))
    for i in range(N):
        H[2*i, 2*i+1] = v        # 胞内跳跃
        H[2*i+1, 2*i] = v
        if i < N-1:
            H[2*i+1, 2*(i+1)] = w   # 胞间跳跃
            H[2*(i+1), 2*i+1] = w
    return H

N = 20
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# 普通相 (v > w): 无边缘态
H_trivial = ssh_hamiltonian(v=1.5, w=0.5, N=N)
E_trivial = np.linalg.eigvalsh(H_trivial)
axes[0].scatter(range(len(E_trivial)), np.sort(E_trivial),
               c=['red' if abs(e) < 0.05 else 'steelblue' for e in np.sort(E_trivial)],
               s=30, zorder=5)
axes[0].axhline(0, color='gray', lw=0.8, linestyle='--')
axes[0].set_title('普通相 (v=1.5 > w=0.5)\n无零能边缘态', fontsize=10)
axes[0].set_ylabel('能量 E')
axes[0].set_xlabel('能级编号')

# 拓扑相 (v < w): 有零能边缘态
H_topo = ssh_hamiltonian(v=0.5, w=1.5, N=N)
E_topo = np.linalg.eigvalsh(H_topo)
axes[1].scatter(range(len(E_topo)), np.sort(E_topo),
               c=['red' if abs(e) < 0.05 else 'steelblue' for e in np.sort(E_topo)],
               s=30, zorder=5)
axes[1].axhline(0, color='gray', lw=0.8, linestyle='--')
axes[1].set_title('拓扑相 (v=0.5 < w=1.5)\n有零能边缘态（红点）', fontsize=10)
axes[1].set_ylabel('能量 E')
axes[1].set_xlabel('能级编号')

# 扫描 w/v 比值，观察相变
ratios = np.linspace(0.1, 2.9, 60)
gap = []
for r in ratios:
    H = ssh_hamiltonian(v=1.0, w=r, N=N)
    E = np.sort(np.linalg.eigvalsh(H))
    gap.append(abs(E[N]) - abs(E[N-1]))  # 最小带隙

axes[2].plot(ratios, gap, 'steelblue', lw=2)
axes[2].axvline(1.0, color='red', linestyle='--', lw=2, label='相变点 w=v')
axes[2].fill_betweenx([0, max(gap)], 0, 1, alpha=0.1, color='blue', label='普通相')
axes[2].fill_betweenx([0, max(gap)], 1, 3, alpha=0.1, color='orange', label='拓扑相')
axes[2].set_xlabel('w/v（胞间/胞内跳跃比）')
axes[2].set_ylabel('能隙大小')
axes[2].set_title('SSH 模型相图\n（能隙在相变点关闭）', fontsize=10)
axes[2].legend(fontsize=8)

plt.suptitle('SSH 模型：最简单的一维拓扑绝缘体\n（Asbóth 书第一章的内容）', fontsize=12)
plt.tight_layout()
plt.show()

print('拓扑相变的标志：能隙关闭然后重新打开')
print('拓扑相中出现零能边缘态（受体-边对应原理保护）')
print('这个「零能态」就是量子计算中 Majorana 费米子的原型模型')
````
