````python
# 用代码验证线性代数理解
import numpy as np

# 矩阵乘法（神经网络每一层的本质）
W = np.array([[0.5, -0.3], [0.2, 0.8], [-0.1, 0.6]])  # 权重矩阵 (3×2)
x = np.array([[1.0], [2.0]])                            # 输入向量 (2×1)
y = W @ x                                               # 输出 (3×1)
print('权重矩阵 W (3×2):')
print(W)
print('\n输入向量 x (2×1):')
print(x.T, '(转置展示)')
print('\n输出 W@x (3×1):')
print(y)
print('\n→ 这就是神经网络一层的计算：Wx + b')

# 链式法则示例
print('\n' + '='*40)
print('链式法则演示（反向传播的核心）：')
print('L = (y - ŷ)²，y = wx + b')
print('dL/dw = dL/dŷ * dŷ/dw')
w, x_val, y_true = 2.0, 3.0, 10.0
y_hat = w * x_val
dL_dyhat = 2 * (y_hat - y_true)   # = -8
dyhat_dw = x_val                   # = 3
dL_dw = dL_dyhat * dyhat_dw       # 链式法则
print(f'w={w}, x={x_val}, y真={y_true}, y预测={y_hat}')
print(f'dL/dw = {dL_dyhat} × {dyhat_dw} = {dL_dw}')
print(f'梯度为负 → w 需要增大来减小损失')
````

### Week 19-20：Transformer 架构（最重要）

| 概念 | 学习要点 |
|------|----------|

**学习资源：**

**完成标志：** 能从零用 PyTorch 实现并训练 MLP + CNN，理解 Transformer 注意力机制

````python
# 演示注意力机制的核心计算（Transformer 的心脏）
import numpy as np

def softmax(x):
    e = np.exp(x - x.max(axis=-1, keepdims=True))
    return e / e.sum(axis=-1, keepdims=True)

def attention(Q, K, V):
    d_k = Q.shape[-1]
    scores = Q @ K.T / np.sqrt(d_k)   # 相似度得分
    weights = softmax(scores)          # 归一化为概率
    return weights @ V, weights

np.random.seed(0)
seq_len, d_k = 4, 8   # 4 个词，每个词 8 维
Q = np.random.randn(seq_len, d_k)
K = np.random.randn(seq_len, d_k)
V = np.random.randn(seq_len, d_k)

output, weights = attention(Q, K, V)

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

im = axes[0].imshow(weights, cmap='Blues', vmin=0, vmax=1)
axes[0].set_xticks(range(seq_len))
axes[0].set_yticks(range(seq_len))
axes[0].set_xticklabels([f'词{i+1}' for i in range(seq_len)])
axes[0].set_yticklabels([f'词{i+1}' for i in range(seq_len)])
for i in range(seq_len):
    for j in range(seq_len):
        axes[0].text(j, i, f'{weights[i,j]:.2f}', ha='center', va='center', fontsize=9)
axes[0].set_title('注意力权重矩阵\n（行=当前词 关注 列=哪个词）', fontsize=10)
plt.colorbar(im, ax=axes[0])

axes[1].bar(range(seq_len), weights[0], color='steelblue', edgecolor='white')
axes[1].set_xticks(range(seq_len))
axes[1].set_xticklabels([f'词{i+1}' for i in range(seq_len)])
axes[1].set_ylabel('注意力权重')
axes[1].set_title('词1 在计算时对各词的关注程度', fontsize=10)
axes[1].set_ylim(0, 0.6)

plt.suptitle('自注意力机制（Self-Attention）核心计算', fontsize=12)
plt.tight_layout()
plt.show()
print('注意力计算：scores = QK^T / √d_k  →  softmax  →  加权求和 V')
````

````python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# ── 演示量子力学中的核心概念 ──
# 1. 量子态用复数向量表示（这是和经典 ML 最大的区别）
print('=== 量子态基础 ===')
# |0> 和 |1> 是计算基底（类比经典的 0 和 1）
ket_0 = np.array([1, 0], dtype=complex)  # |0>
ket_1 = np.array([0, 1], dtype=complex)  # |1>

# 叠加态：|+> = (|0> + |1>) / sqrt(2)
ket_plus = (ket_0 + ket_1) / np.sqrt(2)
print(f'|0> = {ket_0}')
print(f'|1> = {ket_1}')
print(f'|+> = {ket_plus.round(3)}  （等概率叠加态）')
print(f'测量到 |0> 的概率: {abs(ket_plus[0])**2:.2f}')
print(f'测量到 |1> 的概率: {abs(ket_plus[1])**2:.2f}')

# 2. 量子门是幺正矩阵（可逆、保范数）
print('\n=== 量子门（幺正变换）===')
H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)  # Hadamard 门
X = np.array([[0, 1], [1, 0]])                 # Pauli-X（量子非门）
Z = np.array([[1, 0], [0, -1]])                # Pauli-Z
print(f'Hadamard 门 H:\n{H.round(3)}')
print(f'\nH|0> = {(H @ ket_0).round(3)}  → 从确定态变叠加态')
print(f'H|+> = {(H @ ket_plus).round(3)}  → 叠加态变回确定态')
print(f'\n验证幺正性 H†H = I: {np.allclose(H.conj().T @ H, np.eye(2))}')

# 3. 可视化：Bloch 球上的量子态（量子比特的几何表示）
fig = plt.figure(figsize=(13, 4))

# 左图：概率波函数
ax1 = fig.add_subplot(131)
x = np.linspace(-5, 5, 1000)
psi_real = np.exp(-x**2/2) * np.cos(2*x)   # 实部
psi_imag = np.exp(-x**2/2) * np.sin(2*x)   # 虚部
prob = psi_real**2 + psi_imag**2
ax1.plot(x, psi_real, label='实部 Re(ψ)', color='steelblue')
ax1.plot(x, psi_imag, label='虚部 Im(ψ)', color='coral', linestyle='--')
ax1.fill_between(x, 0, prob, alpha=0.2, color='seagreen', label='|ψ|² 概率密度')
ax1.set_title('波函数（复数！）', fontsize=10)
ax1.legend(fontsize=8)
ax1.set_xlabel('位置 x')

# 中图：量子门操作序列
ax2 = fig.add_subplot(132)
states_label = ['初始|0>', '经过H门\n叠加态', '经过X门\n翻转', '经过H门\n干涉']
states = [ket_0, H @ ket_0, X @ H @ ket_0, H @ X @ H @ ket_0]
probs_0 = [abs(s[0])**2 for s in states]
probs_1 = [abs(s[1])**2 for s in states]
x_pos = range(len(states_label))
ax2.bar(x_pos, probs_0, label='P(|0>)', color='steelblue', alpha=0.7)
ax2.bar(x_pos, probs_1, bottom=probs_0, label='P(|1>)', color='coral', alpha=0.7)
ax2.set_xticks(x_pos)
ax2.set_xticklabels(states_label, fontsize=7.5)
ax2.set_ylabel('测量概率')
ax2.set_title('量子门序列改变测量概率', fontsize=10)
ax2.legend(fontsize=8)

# 右图：经典比特 vs 量子比特
ax3 = fig.add_subplot(133)
ax3.axis('off')
table_data = [
    ['特征', '经典比特', '量子比特'],
    ['状态', '0 或 1', '|0>, |1>, 或叠加态'],
    ['表示', '实数 {0,1}', '复向量 α|0>+β|1>'],
    ['门操作', '逻辑门', '幺正矩阵'],
    ['测量', '确定结果', '概率性坍缩'],
    ['多比特', '独立', '可纠缠'],
]
tbl = ax3.table(cellText=table_data[1:], colLabels=table_data[0],
                loc='center', cellLoc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(8)
tbl.scale(1, 1.6)
ax3.set_title('经典 vs 量子', fontsize=10)

plt.suptitle('量子力学核心概念（AI × 量子物理的数学语言）', fontsize=12)
plt.tight_layout()
plt.show()
````

**Week 33-36：变分量子算法（QML 的核心）**

```
变分量子算法的通用框架：

经典参数 θ
    ↓
量子电路 U(θ)  ← 参数化量子门（类比神经网络层）
    ↓
量子测量  →  期望值 <ψ|H|ψ>
    ↓
经典优化器（Adam 等）更新 θ
    ↑─────────────────────┘
```

| 算法 | 用途 | 重要程度 |
|------|------|----------|
| **VQE（变分量子本征解算器）**| 求分子基态能量（量子化学）| ★★★★★ |
| **QAOA**（量子近似优化）| 组合优化问题 | ★★★★☆ |
| **量子神经网络（QNN）**| 参数化电路做分类/回归 | ★★★★☆ |
| **量子核方法** | 量子特征映射 + 经典 SVM | ★★★☆☆ |

````python
# 用 numpy 模拟一个简单的变分量子电路（VQC）
# 实际用 PennyLane，这里展示核心思想

def Rx(theta):
    """绕 X 轴旋转门"""
    c, s = np.cos(theta/2), np.sin(theta/2)
    return np.array([[c, -1j*s], [-1j*s, c]])

def Ry(theta):
    """绕 Y 轴旋转门"""
    c, s = np.cos(theta/2), np.sin(theta/2)
    return np.array([[c, -s], [s, c]])

def measure_Z(state):
    """测量 Z 方向的期望值 <ψ|Z|ψ>"""
    Z = np.array([[1, 0], [0, -1]])
    return np.real(state.conj() @ Z @ state)

# 变分电路：U(θ) = Ry(θ1) Rx(θ2) |0>
def vqc(params):
    state = np.array([1, 0], dtype=complex)  # |0>
    state = Ry(params[0]) @ state
    state = Rx(params[1]) @ state
    return measure_Z(state)

# 目标：调整参数使期望值接近 -1（即 |1> 态）
target = -1.0
params = np.array([0.1, 0.1])
lr = 0.3
history = []

for step in range(60):
    exp_val = vqc(params)
    loss = (exp_val - target) ** 2
    history.append(loss)
    # 参数位移规则（量子梯度计算方法）
    grad = np.zeros(2)
    for i in range(2):
        p_plus  = params.copy(); p_plus[i]  += np.pi/2
        p_minus = params.copy(); p_minus[i] -= np.pi/2
        grad[i] = (vqc(p_plus) - vqc(p_minus)) / 2 * 2 * (exp_val - target)
    params -= lr * grad

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(history, 'o-', color='steelblue', ms=4)
axes[0].set_xlabel('优化步数')
axes[0].set_ylabel('损失值')
axes[0].set_title('变分量子电路（VQC）训练过程\n（和经典神经网络训练完全类似！）')
axes[0].set_yscale('log')

# 扫描参数空间（量子损失地形）
theta_range = np.linspace(-np.pi, np.pi, 50)
loss_surface = np.array([[((vqc([t1, t2]) - target)**2)
                           for t2 in theta_range] for t1 in theta_range])
im = axes[1].contourf(theta_range, theta_range, loss_surface, levels=20, cmap='RdYlGn_r')
axes[1].plot(params[1], params[0], 'w*', ms=15, label='最优参数')
plt.colorbar(im, ax=axes[1])
axes[1].set_xlabel('θ₂ (Rx 角度)')
axes[1].set_ylabel('θ₁ (Ry 角度)')
axes[1].set_title('量子损失地形（参数空间）')
axes[1].legend()

plt.suptitle('变分量子算法：量子版的梯度下降', fontsize=12)
plt.tight_layout()
plt.show()
print(f'最终期望值: {vqc(params):.4f}（目标值: {target}）')
````

````python
# 演示：用经典 ML 检测量子相变（C2 路线的入门级应用）
# 伊辛模型（Ising Model）是量子多体物理的标准入门模型

# 模拟不同温度下的伊辛模型自旋构型
def generate_ising_configs(T, n_spins=16, n_samples=200):
    configs = []
    for _ in range(n_samples):
        # 简化模拟：低温有序（铁磁），高温无序（顺磁）
        if T < 2.27:  # 伊辛临界温度 Tc ≈ 2.27（2D）
            # 低温：大多数自旋对齐
            spins = np.random.choice([-1, 1], n_spins,
                                     p=[0.5 - 0.4*(2.27-T)/2.27,
                                        0.5 + 0.4*(2.27-T)/2.27])
        else:
            # 高温：随机无序
            spins = np.random.choice([-1, 1], n_spins)
        configs.append(spins)
    return np.array(configs)

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

# 生成数据：低温（有序相）vs 高温（无序相）
T_low  = [0.5, 1.0, 1.5, 1.8]
T_high = [2.8, 3.5, 4.0, 5.0]

X_ordered   = np.vstack([generate_ising_configs(T, n_samples=100) for T in T_low])
X_disordered = np.vstack([generate_ising_configs(T, n_samples=100) for T in T_high])

X_all = np.vstack([X_ordered, X_disordered])
y_all = np.array([0]*len(X_ordered) + [1]*len(X_disordered))  # 0=有序, 1=无序

X_tr, X_te, y_tr, y_te = train_test_split(X_all, y_all, test_size=0.2, random_state=42)
clf = MLPClassifier(hidden_layer_sizes=(32, 16), max_iter=500, random_state=42)
clf.fit(X_tr, y_tr)
print(f'相变检测准确率: {clf.score(X_te, y_te):.2%}')

# 在连续温度上预测相：找到相变点
T_range = np.linspace(0.5, 5.0, 40)
order_params = []
for T in T_range:
    configs = generate_ising_configs(T, n_samples=100)
    prob_disordered = clf.predict_proba(configs)[:, 1].mean()
    order_params.append(prob_disordered)

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# ML 预测的相图
axes[0].plot(T_range, order_params, 'o-', color='steelblue', ms=5)
axes[0].axvline(2.27, color='red', linestyle='--', lw=2, label='理论临界温度 Tc=2.27')
axes[0].axhline(0.5, color='gray', linestyle=':', alpha=0.7)
axes[0].set_xlabel('温度 T')
axes[0].set_ylabel('ML 预测为无序相的概率')
axes[0].set_title('神经网络检测伊辛模型相变\n（AI for 量子物理的入门应用）')
axes[0].legend()

# 不同温度的自旋构型可视化
temps_to_show = [0.5, 2.0, 2.27, 4.0]
for j, T in enumerate(temps_to_show):
    config = generate_ising_configs(T, n_samples=1)[0]
    axes[1].barh(j, sum(config==1)/len(config), color='coral', label='自旋+1' if j==0 else '')
    axes[1].barh(j, -sum(config==-1)/len(config), color='steelblue', label='自旋-1' if j==0 else '')

axes[1].set_yticks(range(4))
axes[1].set_yticklabels([f'T={T}' for T in temps_to_show])
axes[1].set_xlabel('自旋比例')
axes[1].set_title('不同温度的自旋构型\n（低温有序 → 高温无序）')
axes[1].legend(fontsize=9)
axes[1].axvline(0, color='black', lw=0.5)

plt.suptitle('伊辛模型相变检测：用 ML 发现量子相变', fontsize=12)
plt.tight_layout()
plt.show()
````
