````python
# ── 示范：从零用 NumPy 计算石墨烯能带结构 ──
# 这是拓扑物理计算的基础技能，第 2-3 个月应该能独立写出这样的代码

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

def graphene_hamiltonian(kx, ky, t=1.0):
    """石墨烯 2×2 紧束缚哈密顿量"""
    a = 1.0  # 晶格常数
    # 石墨烯三个最近邻跳跃向量
    d1 = np.array([a, 0])
    d2 = np.array([-a/2,  a*np.sqrt(3)/2])
    d3 = np.array([-a/2, -a*np.sqrt(3)/2])
    k = np.array([kx, ky])
    f = t * (np.exp(1j*np.dot(k,d1)) +
             np.exp(1j*np.dot(k,d2)) +
             np.exp(1j*np.dot(k,d3)))
    return np.array([[0, f], [f.conj(), 0]])

# 沿高对称路径 Γ-K-M-Γ 计算能带
a = 1.0
G = np.array([0, 0])
K = np.array([2*np.pi/(3*a), 2*np.pi/(3*np.sqrt(3)*a)])
M = np.array([np.pi/a, np.pi/(np.sqrt(3)*a)])

n_pts = 100
path = np.vstack([
    np.linspace(G, K, n_pts),
    np.linspace(K, M, n_pts),
    np.linspace(M, G, n_pts),
])

energies = []
for kpt in path:
    H = graphene_hamiltonian(kpt[0], kpt[1])
    E = np.linalg.eigvalsh(H)
    energies.append(E)
energies = np.array(energies)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

x = np.arange(len(path))
axes[0].plot(x, energies[:, 0], 'steelblue', lw=2, label='价带')
axes[0].plot(x, energies[:, 1], 'coral', lw=2, label='导带')
axes[0].axhline(0, color='gray', lw=0.8, linestyle='--')
for xv in [0, n_pts, 2*n_pts, 3*n_pts-1]:
    axes[0].axvline(xv, color='black', lw=0.8)
axes[0].set_xticks([0, n_pts, 2*n_pts, 3*n_pts-1])
axes[0].set_xticklabels(['Γ', 'K', 'M', 'Γ'])
axes[0].set_ylabel('能量 E/t')
axes[0].set_title('石墨烯能带结构\n（狄拉克点在 K 处）')
axes[0].legend()

# 狄拉克点附近的线性色散（拓扑性质的体现）
kx_range = np.linspace(-0.5, 0.5, 100)
ky_range = np.linspace(-0.5, 0.5, 100)
KX, KY = np.meshgrid(kx_range, ky_range)
gap = np.zeros_like(KX)
for i in range(len(kx_range)):
    for j in range(len(ky_range)):
        H = graphene_hamiltonian(K[0]+kx_range[i], K[1]+ky_range[j])
        E = np.linalg.eigvalsh(H)
        gap[j,i] = E[1] - E[0]

im = axes[1].contourf(KX, KY, gap, levels=20, cmap='RdYlGn')
plt.colorbar(im, ax=axes[1])
axes[1].plot(0, 0, 'r*', ms=15, label='狄拉克点（能隙=0）')
axes[1].set_xlabel('Δkx')
axes[1].set_ylabel('Δky')
axes[1].set_title('K 点附近的能隙（狄拉克锥）')
axes[1].legend()

plt.suptitle('石墨烯：用 NumPy 计算能带（第 2-3 个月可以做到）', fontsize=11)
plt.tight_layout()
plt.show()
````

````python
# ── 示范：用 kwant 风格（纯 NumPy 模拟）计算量子线电导 ──
# 真实使用时换成 import kwant，这里展示核心计算逻辑

def transmission_1d(E, t=1.0, V0=0.3, N_barrier=5):
    """计算一维量子线穿过势垒的透射率（Transfer Matrix 方法）"""
    # 势垒区域的转移矩阵
    def transfer_matrix(E, V, t):
        eps = E - V  # 有效能量
        if abs(eps) < 2*t:
            k = np.arccos(-eps/(2*t))  # 传播模式
            m = np.array([[np.exp(1j*k), 0],
                           [0, np.exp(-1j*k)]])
        else:
            kappa = np.arccosh(abs(eps)/(2*t))  # 衰减模式
            m = np.array([[np.exp(kappa), 0],
                           [0, np.exp(-kappa)]])
        return m

    # Landauer 公式：G = (e²/h) * T
    # 简化计算：均匀势垒区域
    k_lead = np.arccos(-E/(2*t)) if abs(E) < 2*t else None
    if k_lead is None:
        return 0.0
    # 势垒区有效衰减
    eps_barrier = E - V0
    if abs(eps_barrier) < 2*t:
        T = 1.0  # 共振隧穿
    else:
        kappa = np.arccosh(min(abs(eps_barrier)/(2*t), 50))
        T = 1.0 / (1.0 + (np.sinh(kappa*N_barrier))**2)
    return float(np.real(T))

# 计算不同能量下的电导
E_range = np.linspace(-2.5, 2.5, 300)
T_no_barrier   = [1.0 if abs(E) < 2 else 0 for E in E_range]
T_with_barrier = [transmission_1d(E, V0=0.5, N_barrier=8) for E in E_range]
T_resonant     = [transmission_1d(E, V0=0.8, N_barrier=3) for E in E_range]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].plot(E_range, T_no_barrier,   'steelblue', lw=2, label='无势垒（完美导线）')
axes[0].plot(E_range, T_with_barrier, 'coral',     lw=2, label='有势垒（隧穿）')
axes[0].plot(E_range, T_resonant,     'seagreen',  lw=2, linestyle='--', label='薄势垒（共振）')
axes[0].set_xlabel('费米能量 E/t')
axes[0].set_ylabel('透射率 T（= 电导 × h/e²）')
axes[0].set_title('Landauer 公式：G = (e²/h) × T\n这是 kwant 计算的核心量')
axes[0].legend(fontsize=8)
axes[0].set_ylim(-0.05, 1.2)

# 拓扑边缘态的特征：T=1 的量子化平台
E_topo = np.linspace(-3, 3, 300)
G_normal  = np.where(abs(E_topo) < 1.0, 0.05, np.where(abs(E_topo) > 1.5, 1.0, 0.0))
G_topo    = np.where(abs(E_topo) < 1.0, 1.0, np.where(abs(E_topo) > 1.5, 2.0, 1.0))

axes[1].plot(E_topo, G_normal, 'steelblue', lw=2, label='普通绝缘体（带隙内 G≈0）')
axes[1].plot(E_topo, G_topo,   'coral',     lw=2.5, label='拓扑绝缘体（带隙内 G=e²/h）')
axes[1].axhspan(0.9, 1.1, alpha=0.1, color='red')
axes[1].text(0, 1.15, '量子化平台 G = e²/h\n（拓扑边缘态的特征）',
            ha='center', fontsize=9, color='red')
axes[1].set_xlabel('费米能量 E/t')
axes[1].set_ylabel('电导 G (e²/h)')
axes[1].set_title('拓扑绝缘体 vs 普通绝缘体的输运特征')
axes[1].legend(fontsize=8)

plt.suptitle('量子输运计算：kwant 做的就是这类计算', fontsize=11)
plt.tight_layout()
plt.show()
````

# 典型研究计算工作流

理解这个工作流，就理解了你要用 Python 做什么。

```
1. 定义物理模型
   写出哈密顿量 H(k) 或实空间 H
   工具：NumPy 矩阵 / pythtb 建模

2. 能带计算
   对布里渊区的 k 点求 H(k) 的本征值
   画出 E(k) 能带图
   工具：NumPy.linalg + Matplotlib

3. 判断拓扑性质
   计算 Berry 相位、陈数、Z₂ 不变量
   工具：Z2Pack / pythtb

4. 计算输运性质
   建立带边界的有限系统
   计算散射矩阵和透射率 T
   得到电导 G = (e²/h) × T
   工具：kwant

5. 扫描参数空间，画相图
   改变模型参数（跳跃强度、化学势）
   重复步骤 2-4，记录结果
   工具：Python 循环 + 并行计算（joblib）

6. AI 加速（线 C 开始后）
   用步骤 5 的数据训练 ML 模型
   新参数点直接用模型预测，不用跑完整计算
   加速比：10x ~ 1000x
```

**第 6 步就是 AI 的价值所在：步骤 5 的参数扫描可能需要跑几天，ML 预测只需要几秒。**
