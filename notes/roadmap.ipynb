## 第二部分：六个层次详解

### 第一层：数学基础

**为什么是地基：** 量子力学用线性代数的语言写成，拓扑不变量是微分几何的积分，对称性分类用群论描述。数学不扎实，后面的物理只能死记硬背，无法真正理解。

| 模块 | 具体内容 | 书目 | 关键程度 |
|------|----------|------|----------|
| 线性代数 | 矩阵运算、特征值、SVD、酉矩阵、厄米矩阵 | Strang《线性代数》| ★★★★★ |
| 微积分与复分析 | 链式法则、多元偏导、复数、解析函数基础 | 工程教材复习即可 | ★★★★★ |
| 概率论 | 条件概率、随机变量、正态分布、贝叶斯 | Blitzstein《概率论》| ★★★★☆ |
| 微分几何入门 | 流形直觉、纤维丛（只需物理直觉）| Nakahara 选读 | ★★★★☆ |
| 群论基础 | 群的定义、表示论基础、晶体点群 | Dresselhaus 选读 | ★★★☆☆ |

**学习建议：** 线代和微积分立刻开始。微分几何和群论等到第三层之前再补，不要一开始就啃。

---

### 第二层：核心物理

**为什么重要：** 量子力学是一切的语言，固体物理是这个方向的舞台，统计力学让你理解有限温度效应。缺少任何一个，研究都是残缺的。

| 模块 | 书目 | 必读章节 | 关键程度 |
|------|------|----------|----------|
| 量子力学 | Griffiths《量子力学导论》| **Ch.1-5 全部，做习题** | ★★★★★ |
| 固体物理 | Kittel《固体物理导论》| Ch.1-2（晶体），Ch.7-8（能带），Ch.10（费米气体）| ★★★★★ |
| 统计力学 | Huang《Statistical Mechanics》或 Pathria | 基础热统，费米-狄拉克分布 | ★★★★☆ |

**关键里程碑：** 能独立推导无限深势阱、谐振子的定态解；能解释为什么固体有能带结构。
