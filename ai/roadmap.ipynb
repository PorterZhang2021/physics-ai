> 从零基础出发，建立扎实的 AI/ML 理论与工程能力，

**需要建立的四类能力：**
```

```

这四类能力是顺序依赖的：数学 → 编程 → ML → AI×物理。

1. **知识体系全景图**：各层知识的内容与依赖关系

## 第二部分：六个层次详解

### 第一层：数学基础

**为什么是地基：** 神经网络本质是矩阵运算；梯度下降需要微积分；概率论支撑所有不确定性建模。
数学不扎实，后面的一切都只能死记公式，无法真正理解。

| 模块 | 核心内容 | 书目/资源 | 关键程度 |
|------|----------|-----------|----------|
| **线性代数** | 向量、矩阵运算、特征值分解、SVD | Strang《Introduction to Linear Algebra》Ch.1-2,4,6 | ★★★★★ |
| **微积分** | 导数、偏导、链式法则、梯度 | 3Blue1Brown《Essence of Calculus》视频（免费）+任意工程教材 | ★★★★★ |
| **概率论与统计** | 条件概率、分布、期望、贝叶斯定理 | Blitzstein《Introduction to Probability》Ch.1-5（免费 PDF）| ★★★★☆ |
| **信息论** | 熵、KL 散度、互信息 | 《Mathematics for Machine Learning》Ch.6（免费 PDF）| ★★★☆☆ |

**学习建议：** 线代和微积分立刻并行开始，不要等一门全部学完再开始另一门。
概率论等开始学 ML 前再补。信息论按需补充，不是阻塞项。

**关键里程碑：** 能手推矩阵乘法；能手算 2x2 矩阵特征值；能推导链式法则例子。

| **NumPy** | ndarray、广播机制、矩阵运算、索引切片 | NumPy 官方 Quickstart + 《Python 数据分析》Ch.4 | ★★★★★ |
| **Matplotlib** | 折线图、散点图、热力图、子图 | Matplotlib 官方 Gallery（按需查） | ★★★★☆ |
| **Pandas** | DataFrame、数据清洗、groupby、merge | 《Python for Data Analysis》Ch.5-10 | ★★★☆☆ |
| **工程习惯** | git 版本管理、Jupyter 使用、虚拟环境 | git 官方文档；自行实践 | ★★★☆☆ |

**学习建议：** Python 基础和 NumPy 要扎实，其余工具边用边学。
写代码是学习 Python 的唯一方式，每天至少写 30 分钟代码。

**关键里程碑：** 能用 NumPy 实现矩阵乘法和特征值计算；能用 Matplotlib 画出函数图像。

### 第三层：机器学习基础

**为什么重要：** ML 是 AI 的核心方法论，深度学习是其特例。
不理解 ML 基础（过拟合、交叉验证、特征工程），在深度学习中会犯很多低级错误。

**三大范式：**

| 范式 | 有无标签 | 核心算法 | 在 cross/ 中的应用 |
|------|----------|----------|--------------------|
| **监督学习** | 有（输入+答案）| 线性/逻辑回归、决策树、随机森林、SVM | 拓扑相分类、能带性质预测 |
| **无监督学习** | 无 | K-Means、PCA、自编码器 | 材料特征降维、聚类分析 |
| **强化学习** | 奖励信号 | Q-Learning、PPO | 了解即可（RLHF 背景知识）|

| 模块 | 核心概念 | 资源 | 关键程度 |
|------|----------|------|----------|
| **监督学习算法** | 线性回归到随机森林，理解每个算法的假设 | Geron《Hands-On ML》Ch.1-7（核心）| ★★★★★ |
| **模型评估** | 训练集/验证集/测试集，k 折交叉验证，ROC/AUC | Geron Ch.3 | ★★★★★ |
| **偏差-方差** | 欠拟合/过拟合，正则化（L1/L2），模型选择 | Geron Ch.4 | ★★★★★ |
| **无监督学习** | K-Means、PCA（理解为特征压缩）| Geron Ch.8-9 | ★★★☆☆ |

**工具：** scikit-learn（参考 `ml-fundamentals.ipynb`）

**关键里程碑：** 能用 sklearn 训练一个分类器，画出混淆矩阵和 ROC 曲线，解释为什么过拟合。
