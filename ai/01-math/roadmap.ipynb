## Key Exercises for the AI Side (Different from cross/ Emphasis)

### Linear Algebra (AI Perspective)
- Use matrix multiplication to express neural network forward propagation: `y = W₂ · ReLU(W₁ · x + b₁) + b₂`
- Understand broadcasting: what happens in NumPy with `(3,) + (3,1)`
- Relationship between SVD and PCA: what does dimensionality reduction do

### Calculus (AI Perspective)
- Derive by hand the gradient of `L = (y - ŷ)²` w.r.t. weight `w` (MSE Loss backpropagation)
- Understand why `∂L/∂w = ∂L/∂ŷ · ∂ŷ/∂w` (this is the chain rule)
- Derive the derivative of the sigmoid function by hand

### Probability (AI Perspective)
- Derive cross-entropy loss from maximum likelihood estimation (MLE)
- Understand the meaning and computation of KL divergence `KL(P||Q)`
- Application of Bayes theorem in classification problems (Naive Bayes)

## Information Theory (Learn as needed, stored in info-theory/)

**Trigger timing:** Before starting ML basics (Layer 3), spend 3–5 days for a quick pass.

| Concept | Formula | Use in AI |
|------|------|-------------|
| Entropy | H(X) = -Σ p(x) log p(x) | Information gain in decision trees |
| Cross-Entropy | H(P,Q) = -Σ p(x) log q(x) | Loss function for classification tasks |
| KL Divergence | KL(P\|\|Q) = Σ p(x) log(p/q) | VAE; NQS variational objective |
| Mutual Information | I(X;Y) = H(X) - H(X\|Y) | Feature selection; information bottleneck |

**Resource:** *Mathematics for Machine Learning* (Deisenroth) Ch.6 (free PDF)
