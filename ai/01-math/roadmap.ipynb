## AI 侧的关键练习（区别于 cross/ 的偏重点）

### 线性代数（AI 视角）
- 用矩阵乘法表示神经网络的前向传播：`y = W₂ · ReLU(W₁ · x + b₁) + b₂`
- 理解广播机制（broadcasting）：NumPy 里 `(3,) + (3,1)` 会发生什么
- SVD 和 PCA 的关系：降维是在做什么

### 微积分（AI 视角）
- 手推 `L = (y - ŷ)²` 对权重 `w` 的梯度（MSE Loss 反向传播）
- 理解为什么 `∂L/∂w = ∂L/∂ŷ · ∂ŷ/∂w`（这就是链式法则）
- 手推 sigmoid 函数的导数

### 概率论（AI 视角）
- 从最大似然估计（MLE）推导出交叉熵损失
- 理解 KL 散度 `KL(P||Q)` 的含义和计算
- 贝叶斯定理在分类问题里的应用（朴素贝叶斯）

## 信息论（按需学，存放在 info-theory/）

**触发时机：** 开始学 ML 基础（Layer 3）之前，花 3–5 天快速过一遍。

| 概念 | 公式 | AI 中的用途 |
|------|------|-------------|
| 熵 | H(X) = -Σ p(x) log p(x) | 决策树的信息增益 |
| 交叉熵 | H(P,Q) = -Σ p(x) log q(x) | 分类任务的损失函数 |
| KL 散度 | KL(P\|\|Q) = Σ p(x) log(p/q) | VAE；NQS 变分目标 |
| 互信息 | I(X;Y) = H(X) - H(X\|Y) | 特征选择；信息瓶颈 |

**资源：** 《Mathematics for Machine Learning》(Deisenroth) Ch.6（免费 PDF）
